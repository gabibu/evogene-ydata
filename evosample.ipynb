```python
def join_files(features_file_path, columns_files_path, row_file_path, rows_column_name,  nrows):
    print('load npz')
    feattures_sparse_matrix = sparse.load_npz(features_file_path)
    print('todense')
    features_array = feattures_sparse_matrix.todense()[0:nrows]
    
    columns_df = pd.read_csv(columns_files_path, names=['feature'])
    print('columns_df')
    
    columns = columns_df['feature'].tolist()
    print('columns')
    
    index = pd.Index(range(features_array.shape[0]))
    features_df = pd.DataFrame(data=features_array, index=index, columns=columns)
    print('features_df')
    rows_df = pd.read_csv(row_file_path, names=[rows_column_name],  nrows=nrows)
    print('rows_df')
    rows_df.index = range(features_array.shape[0])
    all_data_df = pd.merge(rows_df, features_df, left_index=True, right_index=True)
    print('merge')
    
    return all_data_df
```

```python
cluster_to_go_df = join_files("evodata/cluster2go_data.npz", 'evodata/cluster2go_data.cols.txt', 'evodata/cluster2go_data.rows.txt',
          'gene', nrows)
```
